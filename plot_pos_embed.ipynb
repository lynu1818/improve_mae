```python
import torch
from matplotlib import pyplot as plt
import numpy as np
```

```python
state_dict = torch.load('logs/mae_hiera_tiny_224/epoch-epoch=399.ckpt', map_location='cpu')
#state_dict = torch.load('logs/in1k_mae/mae_hiera_base_plus_512/last.ckpt', map_location='cpu')

pos_embed = state_dict['state_dict']['model.pos_embed']


#state_dict = torch.load('official_ckpt/mae_hiera_tiny_224.pth', map_location='cpu')
#pos_embed = state_dict['model_state']['pos_embed']
```

```python
pos_embed_ch = []
for c in range(pos_embed.shape[2]):
    tmp = pos_embed[0, :, c].detach().numpy()
    size = int(np.sqrt(tmp.shape[0]))
    pos_embed_ch.append(tmp.reshape(size, size))
```

```python

for idx, p in enumerate(pos_embed_ch):
    plt.imshow(p, cmap='hot', interpolation='nearest')
    plt.title(f'Channel {idx}')
    plt.show()
    if idx > 10:
        break
```
